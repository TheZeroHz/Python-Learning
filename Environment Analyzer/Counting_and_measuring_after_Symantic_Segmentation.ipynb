**Defining the masking colors according to the classes with** (BGR color format)


````python
building_mask_color=(152,16,60)
tree_mask_color=(58,221,254)
water_mask_color=(41,169,226)
land_mask_color=(246,41,132)
road_mask_color=(228,193,110)
bridge_mask_color=road_mask_color
````

`**Binary Mask image genarating function from a specified mask class**`

````python
def create_binary_mask(masked_image, target_color):
    lower_color = np.array(target_color, dtype=np.uint8)
    upper_color = np.array(target_color, dtype=np.uint8)
    x=cv2.inRange(masked_image, lower_color, upper_color)
    return x
````

### **An example:  Read an image and find out the binary masking for water class**

````python
img_loc='/content/drive/MyDrive/Colab Notebooks/datasets/satelite/DubiaDataset/Tile 1/images/image_part_008.jpg'
masks_loc='/content/drive/MyDrive/Colab Notebooks/datasets/satelite/DubiaDataset/Tile 1/masks/image_part_008.png'
img=cv2.imread(img_loc)
masked_img=cv2.imread(masks_loc)
water_binary_mask=create_binary_mask(masked_img, road_mask_color)
cv2_imshow(water_binary_mask)
````

## **Drawing contours from specified binary masked image**

````python
edges = cv2.Canny(water_binary_mask, 100, 100)
dilated_edges = cv2.dilate(edges, None, iterations=1)
contours, _ = cv2.findContours(dilated_edges.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
contoured_mask = np.zeros_like(water_binary_mask)
cv2.drawContours(contoured_mask, contours, -1, (255, 193, 0), 1)
cv2_imshow(contoured_mask)
````




## **Draw contours on Original Image and Count number of regions**

````python
num_region=0
for contour in contours:
  contoured_img=cv2.drawContours(img, [contour], -1, water_mask_color, 2)
  num_region=num_region+1
print('Total water region is approx: ',num_region)
cv2_imshow(contoured_img)
````

# Finding the **Area and Percentage** from specific binary mask class

> Indented block



````python
total_area=1000*1000 #Here let's assume total actual area is 1000000 KM^2 
total_pixels=img.shape[0]*img.shape[1]
white_pixels = cv2.countNonZero(water_binary_mask) #image from previous example
class_percent =(white_pixels/total_pixels)*100
class_area =(total_area/total_pixels)*white_pixels
print('Area of Water= ',class_area,' km^2')
print('Percentage of Water= ',class_percent,' %')
````
